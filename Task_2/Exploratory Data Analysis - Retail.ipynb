```python
df.groupby('Segment')[['Profit', 'Sales']].sum().plot.bar(color = ['red', 'black'], figsize = (8, 5))
plt.ylabel('Profit / Loss and Sales')
plt.show()
```

<h1 style="color:black">Observation 5</h1>

<h4 style="color:black">Profit and Sales are Maximum in Consumer segment and Minimum in Home Office segment.</h4>

```python
plt.figure(figsize = (12, 8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x = 'Region', y = 'Sales', data = df, hue = 'Segment', order = df['Region'].value_counts().index, palette = 'rocket')
plt.xlabel('Region', fontsize = 15)
plt.show()
```

<h1 style="color:black">Observation 6</h1>

<h4 style="color:black">Segment wise sales are almost same in each Region</h4>

```python
df.groupby('Region')[['Profit', 'Sales']].sum().plot.bar(color = ['red', 'black'], figsize = (8, 5))
plt.ylabel('Profit / Loss and Sales')
plt.show()
```

<h1 style="color:black">Observation 7</h1>

<h4 style="color:black">Profit and Sales are Maximum in West Region and Minimum in South Region</h4>

```python
ps = df.groupby('State')[['Sales', 'Profit']].sum().sort_values(by = 'Sales', ascending = False)
ps[:].plot.bar(color = ['red', 'black'], figsize = (15, 8))
plt.title('Profit / Loass and Sales across States')
plt.xlabel('States')
plt.ylabel('Profit / Loss and Sales')
plt.show()
```

<h1 style="color:black">Observation 8</h1>

<h4 style="color:black">High Profit: 1. California 2. New York 3. Washington</h4>

<h4 style="color:black">Loss: 1. Texas 2. Pensylvania 3. Florida</h4>

```python
t_states = df['State'].value_counts().nlargest(10)
t_states
```

```python
df.groupby('Category')[['Profit', 'Sales']].sum().plot.bar(color = ['red', 'black'], alpha = 0.9, figsize = (8, 5))
plt.ylabel('Profit / Loss and Sales')
plt.show()
```

<h1 style="color:black">Observation 9</h1>

<h4 style="color:black">Technology and Office Supplies have High Profit</h4>

<h4 style="color:black">Furniture have less Profit</h4>

```python
ps = df.groupby('Sub-Category')[['Sales', 'Profit']].sum().sort_values(by = 'Sales', ascending = False)
ps[:].plot.bar(color = ['red', 'black'], figsize = (15, 8))
plt.title('Profit / Loss across States')
plt.xlabel('Sub-Category')
plt.ylabel('Profit / Loss and Sales')
plt.show()
```
